```python
#Map :-
#Explain the purpose of the `map()` function in Python and provide an example of how it can be used 
#to apply a function to each element of an iterable.
#Write a Python program that uses the `map()` function to square each element of a list of numbers.

# Python program to demonstrate working
# of map.
 
# Return double of n
def addition(n):
    return n + n

def square(n):
    return n*n
 
# We double all numbers using map()
numbers = (1, 2, 3, 4)
result = map(addition, numbers)
print(list(result))
resultsq = map(square, numbers)
print(list(resultsq))


#How does the `map()` function differ from a list comprehension in Python, 
#and when would you choose one over the other?
#Create a Python program that uses the `map()` function to convert a list of names to uppercase.
strname  = ['swapnil','shivali','shreesha']
def upper(strm):
    return strm.upper()
uppercase = map(upper,strname)
print(list(uppercase))
#Write a Python program that uses the `map()` function to calculate the length of each 
#word in a list of strings.

def countstr(strm):
    return len(strm)
strcount = map(countstr,strname)
print(list(strcount))

#How can you use the `map()` function to apply a custom function to elements of multiple lists 
#simultaneously in Python?

# map() with multiple iterable arguments of different sizes
list_numbers = [1, 2, 3, 4]
tuple_numbers = (5, 6, 7, 8, 9, 10)
map_iterator = map(lambda x, y: x * y, list_numbers, tuple_numbers)
print(list(map_iterator))

map_iterator = map(lambda x, y: x * y, tuple_numbers, list_numbers)
print(list(map_iterator))
#Create a Python program that uses `map()` to convert a list of temperatures 
#from Celsius to Fahrenheit.
#Write a Python program that uses the `map()` function to round each element of a list 
#of floating-point numbers to the nearest integer.
listfloat = [1.20 , 3.75 , 6.80 , 8.90 , 2.10]
print(round(1.20,0))
round1 = lambda x : round(x,0)
print(round1(10.20))
map_rounding = map(lambda x : round(x,0), listfloat)
print(list(map_rounding))
```

```python
#What is the `reduce()` function in Python, and what module should you import to use it? 
#Provide an example of its basic usage.
# python code to demonstrate working of reduce() 
  
# importing functools for reduce() 
import functools 
  
# initializing list 
lis = [1, 3, 5, 6, 2] 
  
# using reduce to compute sum of list 
print("The sum of the list elements is : ", end="") 
print(functools.reduce(lambda a, b: a+b, lis)) 
  

#Write a Python program that uses the `reduce()` function to find the product of all elements in 
#a list.
print("The product  of the list is : ", end="") 
print(functools.reduce(lambda a, b: a * b , lis))
#Create a Python program that uses `reduce()` to find the maximum element in a list of numbers.
# using reduce to compute maximum element from list 
print("The maximum element of the list is : ", end="") 
print(functools.reduce(lambda a, b: a if a > b else b, lis))
#How can you use the `reduce()` function to concatenate a list of strings into a single string?
strlist = ['India' , 'Is' 'My' 'Country']
print("The complete conconcated string is : ", end="") 
print(functools.reduce(lambda a, b: a + b , strlist))
#Write a Python program that calculates the factorial of a number using the `reduce()` function.
n = 4
print("The factorial of the number is : ", end="") 
print(functools.reduce(lambda a, b: a * b , range(1,n+1)))
#Create a Python program that uses `reduce()` to find the GCD (Greatest Common Divisor) 
#of a list of numbers.
#Write a Python program that uses the `reduce()` function to find the sum of the digits of 
#a given number.
n1 = 14567
strn1 = str(n1)

print("The sum of the digit is : ", end="") 
print(functools.reduce(lambda a, b: int(a) + int(b) , strn1))
```
